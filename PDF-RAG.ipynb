## Simple PDF Query

```python
from langchain_pinecone import PineconeVectorStore,Pinecone
import langchain
import chromadb
import langchain_core
import os
from langchain.embeddings import SentenceTransformerEmbeddings
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
from google.colab import userdata
os.environ["PINECONE_API_KEY"] =  userdata.get('PINECONE_API_KEY')
os.environ["GROQ_API_KEY"] = userdata.get('GROQ_API_KEY')
```

```python
loader = PyPDFLoader("/content/drive/MyDrive/Colab_Notebooks/Langchain/Lang Chain/BreedReportEnglish04-08-2022.pdf")
docs = loader.load()
len(docs)
```

```python
#Split the text into characters
splitter = RecursiveCharacterTextSplitter(chunk_size=800,chunk_overlap=200)
chuncked = splitter.split_documents(docs)
len(chuncked)
```

```python
#Embed the characters
embeddings = SentenceTransformerEmbeddings()
vec_store = PineconeVectorStore(index_name = "animal-breed",embedding = embeddings)
#vec_store.add_documents(chuncked)
```

```python
#Now Reading the data from vec_store
from langchain.chains.question_answering import load_qa_chain
from langchain_groq import ChatGroq
```

```python
llm = ChatGroq()
chain = load_qa_chain(llm,chain_type="stuff")
```

```python
query = "How many  Indigenous species of Sheep are mentioned in the document?"
docs = vec_store.similarity_search(query)
chain.run(input_documents=docs,question=query)
```

```python
query = "Name Indigenous species of Goat with highest percentage share?"
docs = vec_store.similarity_search(query)
chain.run(input_documents=docs,question=query)
```

```python
query = "What is percentage share Hassan Sheep of Total?"
docs = vec_store.similarity_search(query)
chain.run(input_documents=docs,question=query)
```
